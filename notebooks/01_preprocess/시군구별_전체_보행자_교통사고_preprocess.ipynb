```python
import pandas as pd
```

```python
df_raw = pd.read_csv('../../data/raw_data/시군구별_보행자_교통사고.csv', header=None, index_col=None)
```

```python
df_raw
```

```python
df_raw[0] = df_raw[0].fillna(method='ffill')
```

```python
df_raw
```

```python
df_raw.loc[1, 1] = df_raw.loc[0, 1]
```

```python
df_raw
```

```python
pd.concat([df_raw.iloc[:, :2], df_raw.iloc[:, 2:8]], axis = 1)
```

```python
dfs = [
    pd.concat([df_raw.iloc[:, :2], df_raw.iloc[:, 2:8]], axis = 1), 
    pd.concat([df_raw.iloc[:, :2], df_raw.iloc[:, 8:14]], axis = 1), 
    pd.concat([df_raw.iloc[:, :2], df_raw.iloc[:, 14:20]], axis = 1), 
    pd.concat([df_raw.iloc[:, :2], df_raw.iloc[:, 20:26]], axis = 1), 
    pd.concat([df_raw.iloc[:, :2], df_raw.iloc[:, 26:32]], axis = 1)
]
```

```python
i=0
for one_df in dfs:
    year = one_df.iloc[0, 2]
    one_df.drop([0], inplace=True)
    one_df.columns = one_df.iloc[0]
    one_df.drop([1], inplace=True)
    one_df.reset_index(drop=True, inplace=True)
    one_df['연도'] = year
    dfs[i] = one_df.reindex(columns=(['연도'] + [col for col in one_df.columns if col != '연도']), )
    i+=1
    
dfs[0]
```

```python
df_result = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
df_result
```

```python
df_result.rename(columns={'(중상자[명])': '중상자[명]'}, inplace=True)
df_result.rename(columns={'(경상자[명])': '경상자[명]'}, inplace=True)
df_result.rename(columns={'(부상신고자[명])': '부상신고자[명]'}, inplace=True)
df_result
```

```python
df_result['시도'] = df_result['시도'].str.replace(' ', '')
df_result['시군구'] = df_result['시군구'].str.replace(' ', '')
```

```python
df_result
```

```python
df_result.to_csv('../../data/preprocessed_data/시군구별_전체_보행자_교통사고_전처리ver.csv', index=False)
```
